```python
from data_loader import load_data
from two_layer import TwoLayerNet

from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
%matplotlib inline
```

### Load Data
The data_loader.py file contains the template program for loading the dataset. The file includes a function named load_data(), which reads the input data from a file and generates the appropriate training and test sets. You need to modify the load_data() function to complete this task.


```python
file_path = '.../BCW-data.csv'
X_train, X_test, Y_train, Y_test = load_data(file_path)
```

### Two layer classifier

#### Test-1: 
A two layer model with sigmoid acitvation function for input layer.

```python
model = TwoLayerNet(input_dim = 9, hidden_dim = 5, epochs =100,learning_rate= 0.01 ,activation_type = 'sigmoid' )
model.fit(X_train, Y_train, plot_loss = True )
Y_hat = model.predict(X_test)
print('Test Accuracy',accuracy_score(Y_test, Y_hat))   
```

#### Test-2:
A two layer model with ReLU acitvation function for input layer.

```python
model = TwoLayerNet(input_dim = 9, hidden_dim = 5, epochs =100 ,activation_type = 'ReLU' )
model.fit(X_train, Y_train, plot_loss = True  )
Y_hat = model.predict(X_test)
print('Test Accuracy',accuracy_score(Y_test, Y_hat))  
```

#### Question:
What is the effect of the hidden_dim on the model accuracy for each activation function?

```python
hidden_dims = [ 5, 10, 20, 30]
ReLU_loss = list()
Sigmoid_loss = list()

index = 0

np.random.seed(1)
for hidden_dim in hidden_dims:
    model = TwoLayerNet(input_dim = 9, hidden_dim = hidden_dim , epochs =100 ,activation_type = 'sigmoid' )
    model.fit(X_train, Y_train, plot_loss = False )
    Y_hat = model.predict(X_test)
    Sigmoid_loss.append(accuracy_score(Y_test, Y_hat))
    
    model = TwoLayerNet(input_dim = 9, hidden_dim = hidden_dim , epochs =100 ,activation_type = 'ReLU' )
    model.fit(X_train, Y_train, plot_loss = False )
    Y_hat = model.predict(X_test)
    ReLU_loss.append(accuracy_score(Y_test, Y_hat))
      
    
plt.plot(hidden_dims, ReLU_loss, 'rv-')
plt.plot(hidden_dims, Sigmoid_loss, 'bo--')
plt.xlabel('hidden layer')
plt.ylabel('Accuracy')
plt.title('Classification')
plt.legend(['ReLU', 'Sigmoid'])
plt.ylim([0.5,1.0])

```
